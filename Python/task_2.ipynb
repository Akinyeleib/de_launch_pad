```python
# --- Data cleaning ---
```

```python
clean_df = df.copy()
```

```python
pd.to_datetime(clean_df['last_review'], errors='coerce')
```

```python
clean_df['last_review'] = pd.to_datetime(clean_df['last_review'], errors='coerce')
```

```python
clean_df.info()
```

```python
# --- Data enrichment ---
# price_per_booking = price * minimum_nights
```

```python
clean_df['price_per_booking'] = clean_df['price'] * clean_df['minimum_nights']
```

```python
clean_df.info()
```

```python
# Zero Listings
clean_df[clean_df['availability_365'] == 0]
```

```python
clean_df = clean_df[clean_df['availability_365'] > 0]
```

```python
# Zero Listings Removal COnfirmation
clean_df[clean_df['availability_365'] == 0].count()
```

```python
# Bucket availability into categories using availability_365
def availability_bucket(x):
    if pd.isna(x):
        return 'Unknown'
    if x > 300:
        return 'Full-time'
    if 100 <= x <= 300:
        return 'Part-time'
    if x < 100:
        return 'Rare'
    return 'Unknown'
```

```python
clean_df['availability_bucket'] = clean_df['availability_365'].apply(availability_bucket)
```

```python
clean_df['availability_bucket'].head(3)
```

```python
top10_neighbour_by_price = (clean_df.groupby('neighbourhood')['price']
                                .mean()
                                .sort_values(ascending=False)
                                .head(10)
                                .reset_index()
                                .rename(columns={'price':'avg_price'}))

top10_neighbour_by_price
```

```python
avg_avail_price_by_room = (clean_df.groupby('room_type')
                               .agg(avg_availability=('availability_365','mean') if 'availability_365' in clean_df.columns else ('price','mean'),
                                    avg_price=('price','mean'))
                               .reset_index())

avg_avail_price_by_room
```

```python
host_counts = clean_df.groupby(['host_id', 'host_name']).size().reset_index(name='listing_count')
host_most_listings = host_counts.sort_values('listing_count', ascending=False).head(1)

host_most_listings
```

```python
avg_price_by_borough = (clean_df.groupby('neighbourhood_group')['price']
                            .mean()
                            .reset_index()
                            .rename(columns={'price':'avg_price'})
                            .sort_values('avg_price', ascending=False))

avg_price_by_borough
```
